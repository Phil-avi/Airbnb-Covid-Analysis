```python
import requests
import urllib.request
from bs4 import BeautifulSoup
import time
import os
import datetime
import re
from urllib.error import HTTPError
# from dask import delayed
# from dask.distributed import Client
```

```python
url='http://insideairbnb.com/get-the-data.html'
response=requests.get(url)
soup=BeautifulSoup(response.text,'html.parser')
```

```python
city_names=['Barossa Valley','Barwon South West, Vic','Melbourne','Northern Rivers','Sydney','Tasmania']
```

```python
# Scraping the data from Inside Airbnb
for i in soup.find_all('tbody'):
    city=i.tr.find_all('td')[1].string
    if city in city_names:
        if not os.path.exists(city):
            os.mkdir(city)
        for b in i.find_all('tr'):
            if b.a.contents[0]=='neighbourhoods.csv' or b.a.contents[0]=='neighbourhoods.geojson':
                continue
            else:
                date=b.td.string.replace(',','')
                month=datetime.datetime.strptime(date.split(' ')[1], "%B").month
                date_new=date.split(' ')[2]+'-'+str(month)+'-'+date.split(' ')[0]
                url=b.a['href']
                file=b.a.contents[0]
                file_name=date_new+'_'+file
                if os.path.isfile(f'./{city}/{file_name}'):
                    continue
                else:
                    try:
                        urllib.request.urlretrieve(url,f'./{city}/{file_name}')
                        time.sleep(1)
                    except HTTPError as e:
                        print(f'error:{file_name}')
                        continue
    print(city)
```
